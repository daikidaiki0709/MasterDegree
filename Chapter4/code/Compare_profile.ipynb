# プロファイルの描写

### １サンプルのプロファイル
### 各貯蔵期間の平均プロファイル

```python
# 必要モジュールのimport
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 必要なデータのimport
profile_3mm_11mm_df = pd.read_csv('./../data/profile_3mm_11mm.csv',index_col=0)
profile_3mm_36mm_df = pd.read_csv('./../data/profile_3mm_36mm.csv',index_col=0)
profile_sg_3mm_11mm_df = pd.read_csv('./../data/profile_sg_3mm_11mm.csv',index_col=0)
profile_sg_3mm_36mm_df = pd.read_csv('./../data/profile_sg_3mm_36mm.csv',index_col=0)
```

# 1サンプルのプロファイル

```python
fig = plt.figure(figsize=(10,10))

plt.plot(profile_3mm_36mm_df['distance(cm)']*10,
         profile_3mm_36mm_df.iloc[:,1],
         c='k',linewidth=3)
plt.xlim(3,35)
plt.ylim(4.5,7.5)
plt.xlabel('Radial distance [mm]', fontsize=16)
plt.ylabel('the common logarithmic transformed intensity', fontsize=16)
plt.text(3.5,4.6,'(a)',fontsize=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()

fig.savefig('./../output/図_profile_example.jpeg',dpi=400)
```

# 各貯蔵期間の平均プロファイル

```python
# 各週間ごとに分割

apple_smooth_0week = profile_sg_3mm_36mm_df.iloc[:,1:21]
apple_smooth_1week = profile_sg_3mm_36mm_df.iloc[:,21:41]
apple_smooth_2week = profile_sg_3mm_36mm_df.iloc[:,41:61]
apple_smooth_3week = profile_sg_3mm_36mm_df.iloc[:,61:81]
apple_smooth_4week = profile_sg_3mm_36mm_df.iloc[:,81:101]
apple_smooth_5week = profile_sg_3mm_36mm_df.iloc[:,101:121]
apple_smooth_6week = profile_sg_3mm_36mm_df.iloc[:,121:141]
apple_smooth_7week = profile_sg_3mm_36mm_df.iloc[:,141:161]

# 各週間ごとに平均値を算出
apple_smooth_0week_mean = apple_smooth_0week.mean(axis=1)
apple_smooth_1week_mean = apple_smooth_1week.mean(axis=1)
apple_smooth_2week_mean = apple_smooth_2week.mean(axis=1)
apple_smooth_3week_mean = apple_smooth_3week.mean(axis=1)
apple_smooth_4week_mean = apple_smooth_4week.mean(axis=1)
apple_smooth_5week_mean = apple_smooth_5week.mean(axis=1)
apple_smooth_6week_mean = apple_smooth_6week.mean(axis=1)
apple_smooth_7week_mean = apple_smooth_7week.mean(axis=1)
```

```python
# プロット

# 距離を定義
distance = profile_sg_3mm_36mm_df['distance(cm)']*10
fig =plt.figure(figsize=(10,10))
plt.plot(distance,apple_smooth_0week_mean,c='r',label='0weeks',linewidth=3)
plt.plot(distance,apple_smooth_1week_mean,c='orange',label='1weeks',linewidth=3)
plt.plot(distance,apple_smooth_2week_mean,c='yellow',label='2weeks',linewidth=3)
plt.plot(distance,apple_smooth_3week_mean,c='greenyellow',label='3weeks',linewidth=3)
plt.plot(distance,apple_smooth_4week_mean,c='turquoise',label='4weeks',linewidth=3)
plt.plot(distance,apple_smooth_5week_mean,c='dodgerblue',label='5weeks',linewidth=3)
plt.plot(distance,apple_smooth_6week_mean,c='orchid',label='6weeks',linewidth=3)
plt.plot(distance,apple_smooth_7week_mean,c='pink',label='7weeks',linewidth=3)



plt.xlim(3,35)
plt.ylim(4.5,7.5)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Radial distance [mm]', fontsize=16)
plt.ylabel('the common logarithmic transformed intensity', fontsize=16)
plt.legend(fontsize=24)

plt.text(3.5,4.6,'(b)',fontsize=24)
plt.show()

# fig.savefig('0_3week_profile_compare.jpeg')
# fig.savefig('4_7week_profile_compare.jpeg')
fig.savefig('./../output/図_all_week_profile_compare.jpeg',dpi=400)
```
