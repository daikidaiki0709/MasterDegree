## 貯蔵期間ごとの硬さの推移

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

firmness_all = pd.read_csv('./../data/2020_fuji_firmness_mean.csv',index_col=0)

firmness_all_std = pd.read_csv('./../data/2020_fuji_firmness_std.csv',index_col=0)
```

```python
firmness_0week = np.mean(firmness_all.iloc[:20,:])
firmness_1week = np.mean(firmness_all.iloc[20:40,:])
firmness_2week = np.mean(firmness_all.iloc[40:60,:])
firmness_3week = np.mean(firmness_all.iloc[60:80,:])
firmness_4week = np.mean(firmness_all.iloc[80:100,:])
firmness_5week = np.mean(firmness_all.iloc[100:120,:])
firmness_6week = np.mean(firmness_all.iloc[120:140,:])
firmness_7week = np.mean(firmness_all.iloc[140:160,:])

firmness_mean_df = pd.concat([firmness_0week,firmness_1week],axis=0)
firmness_mean_df = pd.concat([firmness_mean_df,firmness_2week],axis=0)
firmness_mean_df = pd.concat([firmness_mean_df,firmness_3week],axis=0)
firmness_mean_df = pd.concat([firmness_mean_df,firmness_4week],axis=0)
firmness_mean_df = pd.concat([firmness_mean_df,firmness_5week],axis=0)
firmness_mean_df = pd.concat([firmness_mean_df,firmness_6week],axis=0)
firmness_mean_df = pd.concat([firmness_mean_df,firmness_7week],axis=0)
firmness_mean_df.columns = 'firmness_mean_eachweek'
firmness_mean_df.index = ['firmness_0week_mean','firmness_1week_mean','firmness_2week_mean',
                          'firmness_3week_mean','firmness_4week_mean','firmness_5week_mean',
                          'firmness_6week_mean','firmness_7week_mean']
firmness_mean_df = firmness_mean_df*9.81

firmness_0week_std = 0.006948374*9.81*np.sqrt(153)
firmness_1week_std = 0.006849059*9.81*np.sqrt(175)
firmness_2week_std = 0.00948732*9.81*np.sqrt(180)
firmness_3week_std = 0.00551499*9.81*np.sqrt(180)
firmness_4week_std = 0.00721616*9.81*np.sqrt(180)
firmness_5week_std = 0.008107164*9.81*np.sqrt(180)
firmness_6week_std = 0.009030991*9.81*np.sqrt(180)
firmness_7week_std = 0.010253268*9.81*np.sqrt(180)

y_err = [firmness_0week_std,firmness_1week_std,firmness_2week_std,
         firmness_3week_std,firmness_4week_std,firmness_5week_std,
         firmness_6week_std,firmness_7week_std]
```

```python
firmness_mean_df[0]
```

```python
sns.set()
sns.set_style('whitegrid')
sns.set_palette('gray')

fig = plt.figure(figsize=(12,8))
x_labels = [i for i in range(8)]
plt.plot(x_labels,firmness_mean_df,linewidth=2,linestyle='dashdot')
plt.scatter(x_labels,firmness_mean_df,s=70)
plt.errorbar(x_labels, firmness_mean_df, yerr = y_err, capsize=5, fmt='o', markersize=10, ecolor='black', markeredgecolor = "black", color='w')
plt.xticks(fontsize=28)
plt.yticks(fontsize=28)
plt.xlabel('Storage Period (week)',fontsize=32)
plt.ylabel('Firmness (N)',fontsize=32)
plt.xlim(-0.5,7.5)
plt.ylim(18,27)

# アルファベットの付与
num = 0.1
ff = 20
hei = 0.3
plt.text(0+num,firmness_mean_df[0]+hei,'a',fontsize=ff)
plt.text(1+num,firmness_mean_df[1]+hei,'b',fontsize=ff)
plt.text(2+num,firmness_mean_df[2]+hei,'c',fontsize=ff)
plt.text(3+num,firmness_mean_df[3]+hei,'d',fontsize=ff)
plt.text(4+num,firmness_mean_df[4]+hei,'d',fontsize=ff)
plt.text(5+num,firmness_mean_df[5]+hei,'e',fontsize=ff)
plt.text(6+num,firmness_mean_df[6]+hei,'f',fontsize=ff)
plt.text(7+num,firmness_mean_df[7]+hei,'g',fontsize=ff)

plt.show()

fig.savefig('./../output/図_firmness_compare.jpeg',dpi=400)
```

## 貯蔵期間ごとの密度の推移

```python
density_0 = [0.881, 0.896, 0.831, 0.878, 0.843]
density_1 = [0.842, 0.810, 0.811, 0.786, 0.837]
density_2 = [0.834, 0.864, 0.847, 0.820, 0.847]
density_3 = [0.85, 0.84, 0.86, 0.87, 0.85]
density_4 = [0.862, 0.824, 0.865, 0.844, 0.859]
density_5 = [0.86, 0.89, 0.85, 0.85]
density_6 = [0.85, 0.86, 0.87, 0.85, 0.84]
density_7 = [0.82, 0.85, 0.82, 0.84, 0.84]

density_list = [density_0, density_1, density_2, density_3,
                density_4, density_5, density_6, density_7]

density_mean = [np.mean(i) for i in density_list]
density_std = [np.std(i) for i in density_list]
```

```python
fig = plt.figure(figsize=(12,8))
x_labels = [i for i in range(8)]
plt.plot(x_labels,density_mean,linewidth=2,linestyle='dashdot')
plt.scatter(x_labels,density_mean,s=70)
plt.errorbar(x_labels, density_mean, yerr = density_std, capsize=5, fmt='o', markersize=10, ecolor='black', markeredgecolor = "black", color='w')
plt.xticks(fontsize=26)
plt.yticks(fontsize=26)
plt.xlabel('Storage Period (week)',fontsize=28)
plt.ylabel('Density (g/cm3)',fontsize=28)
plt.xlim(-0.5,7.5)
plt.ylim(0.78,0.9)
plt.show()

fig.savefig('./../output/図_density_compare.jpeg',dpi=400)
```
